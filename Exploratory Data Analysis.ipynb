```python
import numpy as np
def Process_Size(item):
    if str(item)[-1]=='M':
        res=float(str(item).replace('M',''))
        res=res*1000
        return str(res)
    elif str(item)[-1]=='k':
        res=str(item).replace('k','')
        return res
    else:
        return str(np.NaN)
```
